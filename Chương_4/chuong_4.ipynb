```python
#Câu 4.4
import pandas as pd
ser1=pd.Series([1,2,3,4,5])
ser2=pd.Series([4,5,6,7,8])
result = set(ser1)-set(ser2)
print(result)
```

```python
#Câu 4.5
import pandas as pd
import numpy as np
ser = pd.Series(np.random.normal(10, 5, 25))
min_val = ser.min()
p25 = ser.quantile(0.25)
median = ser.median()
p75 = ser.quantile(0.75)
max_val = ser.max()
print("Min:", min_val)
print("25th Percentile:", p25)
print("Median:", median)
print("", p75)
print("Max:", max_val)
```

```python
#Câu 4.7
np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))
top2 = ser.value_counts().index[:2]
ser[~ser.isin(top2)] = 'Other'
print(ser)
```

```python
#Câu 4.12
ser1 = pd.Series(range(5))
ser2 = pd.Series(list('abcde'))
vertical = pd.concat([ser1, ser2], axis=0)
horizontal = pd.concat([ser1, ser2], axis=1)
print("Vertical Concatenation:\n", vertical)
print("Horizontal Concatenation:\n", horizontal)    
```

```python
#Câu 4.13
ser1 = pd.Series([10, 9, 6, 5, 3, 1, 12, 8, 13])
ser2 = pd.Series([1, 3, 10, 13])
positions = [pd.Index(ser1).get_loc(i) for i in ser2]
print("Vị trí của các phần tử của ser2 trong ser1:", positions)
```

```python
#Câu 4.14
truth = pd.Series(range(10))
pred = pd.Series(range(10)) + np.random.random(10)
mse = np.mean((truth - pred)**2)
print("Mean Squared Error:", mse)
```

```python
#Câu 4.16
ser = pd.Series([1, 3, 6, 10, 15, 21, 27, 35])
diff1 = ser.diff().tolist()
diff2 = ser.diff().diff().tolist()
print("Hiệu số cấp 1:", diff1)
print("Hiệu số cấp 2:", diff2)
```

```python
#Câu 4.18
emails = pd.Series(['buying books at amazon.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}'
result = emails.str.extract(f'({pattern})').dropna()
print(result)
```

```python
#Bài 4.19
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weight = pd.Series(np.linspace(1, 10, 10))
print(weight.tolist())
print(fruit.tolist())
```
