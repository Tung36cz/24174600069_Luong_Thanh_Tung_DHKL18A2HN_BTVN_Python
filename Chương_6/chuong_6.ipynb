```python
#Câu 6.1
import threading
def print_thread_name():
    print(f"Luồng đang chạy: {threading.current_thread().name}")
for i in range(5):
    t = threading.Thread(target=print_thread_name, name=f"Thread-{i+1}")
    t.start()
```

```python
#Câu 6.2
import threading
import time
def download_file(file_name):
    print(f"Bắt đầu tải: {file_name}...")
    time.sleep(2)  
    print(f"Đã tải xong: {file_name}")
files = ["file1.zip", "file2.jpg", "file3.mp4"]
threads = []
for f in files:
    t = threading.Thread(target=download_file, args=(f,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
#Câu 6.3
import threading
def in_so_chan():
    for i in range(30, 51):
        if i % 2 == 0:
            print(f"Số chẵn: {i}")
def in_so_le():
    for i in range(30, 51):
        if i % 2 != 0:
            print(f"Số lẻ: {i}")
t1 = threading.Thread(target=in_so_chan)
t2 = threading.Thread(target=in_so_le)
t1.start()
t2.start()
```

```python
#Câu 6.4
import threading
import math
def tinh_giai_thua(n):
    res = math.factorial(n)
    print(f"Giai thừa của {n} là: {res}")
numbers = [5, 7, 10]
for num in numbers:
    threading.Thread(target=tinh_giai_thua, args=(num,)).start()
```

```python
#Câu 6.5
import threading
import requests
def fetch_url(url):
    response = requests.get(url)
    print(f"URL: {url} - Status Code: {response.status_code}")
urls = ["https://google.com", "https://yahoo.com", "https://facebook.com"]
for url in urls:
    threading.Thread(target=fetch_url, args=(url,)).start()
```

```python
#Câu 6.6
import threading
import time
def viet_so(loai, nguong):
    for i in range(1, nguong + 1):
        if loai == "chan" and i % 2 == 0:
            print(f"Luồng Chẵn: {i}")
        elif loai == "le" and i % 2 != 0:
            print(f"Luồng Lẻ: {i}")
        time.sleep(0.1)
nguong = 10
t_chan = threading.Thread(target=viet_so, args=("chan", nguong))
t_le = threading.Thread(target=viet_so, args=("le", nguong))
t_chan.start()
t_le.start()
```

```python
#Câu  6.7
import threading
import time
from datetime import datetime

def task(name):
    print(f"Starting {name}")
    for _ in range(3):
        now = datetime.now().strftime("%a %b %d %H:%M:%S %Y")
        print(f"{name}: {now}")
        time.sleep(1)
    print(f"Exiting {name}")

threads_name = ["Google", "Yahoo", "Facebook"]
print("Starting Main Thread")
for name in threads_name:
    threading.Thread(target=task, args=(name,)).start()
```

```python
#Câu 6.8
import threading
def sum_sublist(sublist, result_list):
    result_list.append(sum(sublist))
data = list(range(1, 101)) 
mid = len(data) // 2
results = []
t1 = threading.Thread(target=sum_sublist, args=(data[:mid], results))
t2 = threading.Thread(target=sum_sublist, args=(data[mid:], results))
t1.start()
t2.start()
t1.join()
t2.join()

print(f"Tổng cuối cùng: {sum(results)}")
```

```python
#Câu 6.9
import threading
import time

def print_time(threadName, delay, counter):
    print(f"Starting {threadName}")
    while counter:
        time.sleep(delay)
        current_time = time.ctime(time.time())
        print(f"{threadName}: {current_time}")
        counter -= 1
    print(f"Exiting {threadName}")

print("Starting Main Thread")

thread1 = threading.Thread(target=print_time, args=("Google", 1, 3))
thread2 = threading.Thread(target=print_time, args=("Yahoo", 2, 3))
thread3 = threading.Thread(target=print_time, args=("Facebook", 3, 3))

thread1.start()
thread2.start()
thread3.start()

thread1.join()
thread2.join()
thread3.join()

print("Exiting Main Thread")
```

```python
#Câu 6.10
import threading

def find_max(sublist, max_list):
    max_list.append(max(sublist))

nums = [7, 3, 5, 40, 2, 1, 99, 8, 15]
mid = len(nums) // 2
max_vals = []

t1 = threading.Thread(target=find_max, args=(nums[:mid], max_vals))
t2 = threading.Thread(target=find_max, args=(nums[mid:], max_vals))

t1.start()
t2.start()
t1.join()
t2.join()

print(f"Giá trị lớn nhất là: {max(max_vals)}")
```
