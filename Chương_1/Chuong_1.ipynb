```python
#Bai 1
class hinh_chu_nhat:
    def __init__(self,chieu_dai, chieu_rong):
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong
    def chu_vi(self):
        return (self.chieu_dai + self.chieu_rong) * 2
    def dien_tich(self):
        return self.chieu_dai * self.chieu_rong
   
    def ket_qua(self):
        print(f"Chieu dai: {self.chieu_dai}"
              f"\nChieu rong: {self.chieu_rong}"
              f"\nChu vi: {self.chu_vi()}"
              f"\nDien tich: {self.dien_tich()}")
while True:
    Chieu_dai = float(input("Nhap chieu dai: "))
    Chieu_rong = float(input("Nhap chieu rong: "))
    
    if Chieu_dai <= 0 or Chieu_rong <= 0 or Chieu_dai < Chieu_rong:
        print("Chieu dai va chieu rong phai lon hon 0 va chieu dai phai lon hon chieu rong")
        
    else:
        dau_vao = hinh_chu_nhat(Chieu_dai, Chieu_rong)
        break
    
dau_vao.ket_qua()
```

```python
#Bai 3
class PS:
    def __init__(self,tu_so,mau_so):
        self.tu_so = tu_so
        self.mau_so = mau_so
    def kiem_tra(self):
        if self.mau_so == 0:
            return "Mau so phai khac 0"
        else:
            return "Phan so hop le"
    def rut_gon(self):
        def ucln(a,b): 
            while b != 0:
                a, b = b, a % b
            return a
        ucln_value = ucln(abs(self.tu_so), abs(self.mau_so))
        self.tu_so //= ucln_value
        self.mau_so //= ucln_value
        if self.mau_so < 0:
            self.tu_so = -self.tu_so
            self.mau_so = -self.mau_so
    def hien_thi(self):
        print(f"{self.tu_so}/{self.mau_so}")
tu_so = int(input("Nhap tu so: "))
mau_so = int(input("Nhap mau so: "))
phan_so = PS(tu_so,mau_so)
print(phan_so.kiem_tra())
if phan_so.kiem_tra() == "Phan so hop le":
    phan_so.rut_gon()
    print("Phan so sau khi rut gon: ", end="")
    phan_so.hien_thi()
else:
    print("Khong the rut gon phan so vi mau so = 0")
```

```python
#Bai 4
class Stack:
    def __init__(self, size):
        self.size = size
        self.data = [0.0] * size  # mảng float cấp phát động
        self.top = -1

    def __del__(self):
        del self.data

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.size - 1

    def push(self, value):
        if self.isFull():
            print("Stack da day!")
        else:
            self.top += 1
            self.data[self.top] = float(value)

    def pop(self):
        if self.isEmpty():
            print("Stack rong!")
            return None
        else:
            value = self.data[self.top]
            self.top -= 1
            return value
    
s = Stack(5)
s.push(1.5)

print(s.pop())  
print(s.isEmpty())
```

```python
#Bài 7
class Date:
    def __init__(self, day=1, month=1, year=2023):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0)

    def days_in_month(self):
        months = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap_year() and self.month == 2:
            return 29
        return months[self.month]

    def next_day(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

    def display(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

date = Date(28, 2, 2020)
print("Ngay hien tai:", date.display())
```

```python
#Bài 11
class TamGiac:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def tinh_chu_vi(self):
        return self.a + self.b + self.c

    def tinh_dien_tich(self):
        p = self.tinh_chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

class TamGiacCan(TamGiac):
    def __init__(self, canh_ben, canh_day):
        super().__init__(canh_ben, canh_ben, canh_day)

class TamGiacVuong(TamGiac):
    def __init__(self, a, b):
        # c là cạnh huyền
        c = math.sqrt(a**2 + b**2)
        super().__init__(a, b, c)

class TamGiacDeu(TamGiacCan):
    def __init__(self, canh):
        super().__init__(canh, canh)

tg1 = TamGiac(3, 4, 5)
tg2 = TamGiacCan(5, 8)
tg3 = TamGiacVuong(3, 4)
tg4 = TamGiacDeu(6)
print("Tam giac thuong - Chu vi:", tg1.tinh_chu_vi(), "Dien tich:", tg1.tinh_dien_tich())
print("Tam giac can - Chu vi:", tg2.tinh_chu_vi(), "Dien tich:", tg2.tinh_dien_tich())
print("Tam giac vuong - Chu vi:", tg3.tinh_chu_vi(), "Dien tich:", tg3.tinh_dien_tich())
print("Tam giac deu - Chu vi:", tg4.tinh_chu_vi(), "Dien tich:", tg4.tinh_dien_tich())  
```
