```python
#Câu 7.4
import tkinter as tk

def check_entries(*args):
    if ent_name.get() and ent_id.get() and ent_pw.get():
        btn_submit.pack(pady=10)
    else:
        btn_submit.pack_forget()

root = tk.Tk()
root.title("Đăng ký sinh viên")

tk.Label(root, text="Student Name:").pack()
ent_name = tk.Entry(root)
ent_name.pack()
ent_name.bind("<KeyRelease>", check_entries)

tk.Label(root, text="Student ID:").pack()
ent_id = tk.Entry(root)
ent_id.pack()
ent_id.bind("<KeyRelease>", check_entries)

tk.Label(root, text="Password:").pack()
ent_pw = tk.Entry(root, show="*")
ent_pw.pack()
ent_pw.bind("<KeyRelease>", check_entries)

btn_submit = tk.Button(root, text="Submit")

root.mainloop()
```

```python
#Câu 7.5
import tkinter as tk
from PIL import Image, ImageTk
import requests
from io import BytesIO

def load_image_url(url):
    response = requests.get(url)
    img_data = response.content
    img = Image.open(BytesIO(img_data))
    return ImageTk.PhotoImage(img)

root = tk.Tk()
photo = load_image_url("https://example.com/otto.png")
label = tk.Label(root, image=photo)
label.pack()
root.mainloop()
```

```python
#Câu 7.7
import tkinter as tk

def calculate():
    try:
        n = int(ent_n.get())
        total = sum(range(1, n + 1))
        
        lbl_sum.config(text=f"The sum is: {total}")
    except:
        pass

root = tk.Tk()
ent_n = tk.Entry(root)
ent_n.pack()
tk.Button(root, text="Validate", command=calculate).pack()
lbl_sum = tk.Label(root, text="")
lbl_sum.pack()
root.mainloop()
```

```python
#Câu 7.8
import tkinter as tk

def calculate():
    try:
        n = int(ent_n.get())
        divisors = [str(i) for i in range(1, n + 1) if n % i == 0]
        
        lbl_div.config(text=f"The divisors: {' '.join(divisors)}")
    except:
        pass

root = tk.Tk()
ent_n = tk.Entry(root)
ent_n.pack()
tk.Button(root, text="Validate", command=calculate).pack()
lbl_div = tk.Label(root, text="")
lbl_div.pack()
root.mainloop()
```

```python
#Câu 7.9
import tkinter as tk
import math

def solve():
    m, n = int(ent_m.get()), int(ent_n.get())
    gcd = math.gcd(m, n)
    lcm = abs(m * n) // gcd
    lbl_res.config(text=f"GCD: {gcd}, LCM: {lcm}")

root = tk.Tk()
ent_m = tk.Entry(root); ent_m.pack()
ent_n = tk.Entry(root); ent_n.pack()
tk.Button(root, text="Calculate", command=solve).pack()
lbl_res = tk.Label(root, text=""); lbl_res.pack()
root.mainloop()
```

```python
#Câu 7.10
import tkinter as tk

def tinh_tuoi():
    root.title(f"Xin chào {ent_w.get()}")
    name = str(ent_w.get())
    age = float(ent_h.get())
    if age < 18: status = "Bạn còn nhỏ tuổi"
    elif age >= 18: status = "Bạn đã trưởng thành"
    lbl_bmi.config(text=f"Tên: {name} - Tuổi: {age} - {status}")
root = tk.Tk()
ent_w = tk.Entry(root); ent_w.pack()
e_w=tk.Label(root,text="Nhập tên").pack()
ent_h = tk.Entry(root); ent_h.pack()
e_h=tk.Label(root,text="Nhập tuổi").pack()
tk.Button(root, text="Tính tuổi", command=tinh_tuoi).pack()
lbl_bmi = tk.Label(root, text=""); lbl_bmi.pack()
root.mainloop()
```

```python
#Câu 7.12
import tkinter as tk
import math

def tinh_bmi():
    w = float(ent_w.get())
    h = float(ent_h.get())
    bmi = w / (h**2)
    status = "Bình thường"
    if bmi < 18.5: status = "Gầy"
    elif bmi >= 25: status = "Thừa cân"
    lbl_bmi.config(text=f"BMI: {bmi:.2f} - {status}")
root = tk.Tk()
ent_w = tk.Entry(root); ent_w.pack()
e_w=tk.Label(root,text="Nhập cân nặng (kg)").pack()
ent_h = tk.Entry(root); ent_h.pack()
e_h=tk.Label(root,text="Nhập chiều cao (m)").pack()
tk.Button(root, text="Tính BMI", command=tinh_bmi).pack()
lbl_bmi = tk.Label(root, text=""); lbl_bmi.pack()
root.mainloop()
```

```python
#Câu 7.13
import tkinter as tk
from tkinter import messagebox

def show_info(ex_name):
    """Hàm hiển thị thông báo khi chọn một mục menu"""
    if ex_name == "1.1":
        messagebox.showinfo(
            title="Lập trình hướng đối tượng", 
            message="Viết chương trình giải phương trình bậc hai."
        )
    else:
        messagebox.showinfo(
            title=f"Bài tập {ex_name}", 
            message=f"Đây là nội dung của bài tập {ex_name}."
        )

def create_menu_system():
    root = tk.Tk()
    root.title("Hệ thống bài tập Python nâng cao")
    root.geometry("600x400")

    menubar = tk.Menu(root)

    for m in range(1, 8):
        chapter_menu = tk.Menu(menubar, tearoff=0)
        
        for n in range(1, 10):
            ex_label = f"{m}.{n}"
            chapter_menu.add_command(
                label=ex_label, 
                command=lambda name=ex_label: show_info(name)
            )
        
        chapter_menu.add_separator()
        chapter_menu.add_command(label="Exit", command=root.quit)
        
        menubar.add_cascade(label=f"Chương {m}", menu=chapter_menu)

    root.config(menu=menubar)
    root.mainloop()

if __name__ == "__main__":
    create_menu_system()
```

```python
#Câu 7.14
import tkinter as tk
from tkinter import messagebox

def tinh_cuoc():
    try:
        km = float(ent_km.get())
        phut_cho = int(ent_time.get())
        loai_xe = var_xe.get()
        
        tien_cho = 0
        if phut_cho > 5:
            tien_cho = (phut_cho - 5) * 750
        
        if loai_xe == 4:
            gia_duoi_30 = 15300
            gia_tren_31 = 12100
        else:
            gia_duoi_30 = 16100
            gia_tren_31 = 13800
            
        tong_tien_xe = 0
        if km <= 0.8:
            tong_tien_xe = 12000 # Giá mở cửa
        elif km <= 30:
            tong_tien_xe = 12000 + (km - 0.8) * gia_duoi_30
        else:
            tong_tien_xe = 12000 + (29.2 * gia_duoi_30) + (km - 30) * gia_tren_31
            
        tong_cong = tong_tien_xe + tien_cho
        
        lbl_res_xe.config(text=f"Tiền xe: {tong_tien_xe:,.0f} VNĐ")
        lbl_res_cho.config(text=f"Tiền chờ: {tien_cho:,.0f} VNĐ")
        lbl_total.config(text=f"TỔNG CỘNG: {tong_cong:,.0f} VNĐ", fg="red")
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ cho Quãng đường và Thời gian chờ!")

root = tk.Tk()
root.title("Ứng dụng tính cước Taxi")
root.geometry("400x450")

tk.Label(root, text="TÍNH CƯỚC TAXI", font=("Arial", 16, "bold")).pack(pady=10)

tk.Label(root, text="Chọn loại xe:").pack()
var_xe = tk.IntVar(value=4)
tk.Radiobutton(root, text="Xe 4 chỗ", variable=var_xe, value=4).pack()
tk.Radiobutton(root, text="Xe 7 chỗ", variable=var_xe, value=7).pack()

tk.Label(root, text="Quãng đường di chuyển (km):").pack(pady=5)
ent_km = tk.Entry(root)
ent_km.pack()

tk.Label(root, text="Thời gian chờ (phút):").pack(pady=5)
ent_time = tk.Entry(root)
ent_time.insert(0, "0") 
ent_time.pack()

tk.Button(root, text="TÍNH CƯỚC", command=tinh_cuoc, bg="green", fg="white", font=("Arial", 10, "bold")).pack(pady=20)

lbl_res_xe = tk.Label(root, text="Tiền xe: 0 VNĐ")
lbl_res_xe.pack()
lbl_res_cho = tk.Label(root, text="Tiền chờ: 0 VNĐ")
lbl_res_cho.pack()
lbl_total = tk.Label(root, text="TỔNG CỘNG: 0 VNĐ", font=("Arial", 12, "bold"))
lbl_total.pack(pady=10)

root.mainloop()
```
